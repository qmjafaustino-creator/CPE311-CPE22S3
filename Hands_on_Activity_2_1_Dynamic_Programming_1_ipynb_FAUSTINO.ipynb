```python
def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n-1)
print("Factorial using recursion:")

for i in range(7):
    print(f"{i}! = {factorial_recursive(i)}")
```

2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem

```python
def factorial_dp(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"

    dp = [0] * (n + 1)

    dp[0] = 1
    if n >= 1:
        dp[1] = 1

    for i in range(2, n + 1):
        dp[i] = i * dp[i-1]

    return dp[n]

print("Factorial using Dynamic Programming:")
for i in range(7):
    print(f"{i}! = {factorial_dp(i)}")
```

##### Question:
Explain the difference of using the recursion from dynamic programming using the given sample codes to solve the same problem

The recursion. It solves the problem by calling itself repeatedly with smaller inputs until it reaches a base case 0! or 1! = 1. This method is often elegant and easy to understand.In contrast, the dynamic programming uses a bottom-up approach. It builds a table of solutions to subproblems, starting from the base cases. It then iteratively computes using previously calculated values. This way, each subproblem is solved only once, and its result is stored.

```python
def BottomUp(N):
    if N == 0 or N == 1:
        return N

    A = 0  # Changed == to =
    B = 1  # Changed == to =

    for I in range(2, N + 1):  # Corrected for loop syntax
        Temp = A + B  # Changed == to =
        A = B
        B = Temp

    return B

# Example usage (assuming this is meant to be a Fibonacci sequence with bottom-up DP)
print("Fibonacci using Bottom-Up DP:")
for i in range(10):
    print(f"F({i}) = {BottomUp(i)}")
```

4. Create a sample program codes that simulate tops-down dynamic programming

```python
def fib_top_down(n, dp): # Corrected function definition for Python
    # Base case
    if n <= 1:
        return n

    # Check for precomputed result
    if dp[n] != -1:
        return dp[n]

    # Go for subproblem
    # Store the result in dp table before returning
    dp[n] = fib_top_down(n - 1, dp) + fib_top_down(n - 2, dp)
    return dp[n]

# To test the top-down DP (memoization) Fibonacci
def calculate_fibonacci_top_down(n):
    # Initialize dp array with -1 to indicate uncomputed values
    dp = [-1] * (n + 1)
    return fib_top_down(n, dp)

print("Fibonacci using Top-Down DP (Memoization):")
for i in range(10):
    print(f"F({i}) = {calculate_fibonacci_top_down(i)}")
```

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

# initialize the container for the values that have to be stored
# values are initialized to -1
calc = [[-1 for i in range(w+1)] for j in range(n+1)]

def mem_knapSack(wt, val, w, n):
    # base conditions
    if n == 0 or w == 0:
        return 0
    if calc[n][w] != -1:
        return calc[n][w]

    # compute for the other cases
    if wt[n-1] <= w:
        calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                         mem_knapSack(wt, val, w, n-1))
        return calc[n][w]
    elif wt[n-1] > w:
        calc[n][w] = mem_knapSack(wt, val, w, n-1)
        return calc[n][w]

result = mem_knapSack(wt, val, w, n)
print(f": {result}")
```

```python
#type your code here
#Recursion
#sample code for knapsack problem using recursion
def knapsack(capacity, weights, values, n, max_items):
    # Base cases:
    # If no items left or no capacity, or no more items can be picked
    if n == 0 or capacity == 0 or max_items == 0:
        return 0

    # If weight of the current item is more than remaining capacity, it cannot be included
    if weights[n-1] > capacity:
        return knapsack(capacity, weights, values, n-1, max_items)
    else:
        # Option 1: Include the current item
        # Reduce capacity by item's weight, reduce item count by 1
        include = values[n-1] + knapsack(capacity - weights[n-1], weights, values, n-1, max_items - 1)

        # Option 2: Exclude the current item
        # Capacity and item count remain the same
        exclude = knapsack(capacity, weights, values, n-1, max_items)

        # Return the maximum of the two options
        return max(include, exclude)

# Test with different data
values = [60, 100, 120, 80] # Values of items
weights = [5, 10, 15, 7]   # Weights of items
capacity = 25              # Knapsack capacity
n_items = len(values)      # Total number of available items
max_items_allowed = 3      # max items allowed

result_recursive = knapsack(capacity, weights, values, n_items, max_items_allowed)
print(f"Max value (Recursive with max_items): {result_recursive}")


#Dynamic
def DP_knapSack_with_cost(W, C, wt, val, cost, n):
    # Create a 3D DP table: dp[i][w_capacity][c_capacity]
    # i: number of items considered
    # w_capacity: current weight capacity
    # c_capacity: current cost capacity
    dp = [[[0 for _ in range(C + 1)] for _ in range(W + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for w_cap in range(W + 1):
            for c_cap in range(C + 1):
                # Base case: if no items, no weight capacity, or no cost capacity, value is 0
                if i == 0 or w_cap == 0 or c_cap == 0:
                    dp[i][w_cap][c_cap] = 0
                # If current item's weight or cost is more than current capacities, cannot include it
                elif wt[i-1] > w_cap or cost[i-1] > c_cap:
                    dp[i][w_cap][c_cap] = dp[i-1][w_cap][c_cap]
                else:
                    # Option 1: Include the current item
                    # (value of current item + max value from previous items with reduced capacities)
                    include_item = val[i-1] + dp[i-1][w_cap - wt[i-1]][c_cap - cost[i-1]]
                    # Option 2: Exclude the current item
                    # (max value from previous items with same capacities)
                    exclude_item = dp[i-1][w_cap][c_cap]
                    dp[i][w_cap][c_cap] = max(include_item, exclude_item)
    return dp[n][W][C]

# To test Dynamic Programming (max_cost):
val_dp_cost = [60, 100, 120, 80] # Values for the items
wt_dp_cost = [5, 10, 15, 7]   # Weights of the items
cost_dp_cost = [10, 20, 25, 15] # Costs of the items
capacity_dp_cost = 25           # Knapsack weight capacity
max_cost_allowed = 40           # Knapsack cost capacity
num_items_dp_cost = len(val_dp_cost) # Number of available items

result_dp_with_cost = DP_knapSack_with_cost(capacity_dp_cost, max_cost_allowed, wt_dp_cost, val_dp_cost, cost_dp_cost, num_items_dp_cost)
print(f"Max value (Dynamic Programming with max_cost): {result_dp_with_cost}")


#Memoization
val = [60, 100, 120, 80] # Values of items
wt = [10, 20, 30, 7]     # Weights of items
vol = [5, 10, 8, 3]      # NEW: Volumes of the items
W_cap = 50               # Knapsack weight capacity
V_cap = 20               # NEW: Knapsack volume capacity
n = len(val)             # Number of items

#initialize the container for the values that have to be stored
#values are initialized to -1. Now it's a 3D table: memo[item_idx][weight_cap][volume_cap]
memo = [[[-1 for _ in range(V_cap + 1)] for _ in range(W_cap + 1)] for _ in range(n + 1)]


def mem_knapSack_with_volume(wt, vol, val, W_cap, V_cap, n, memo):
  # Base conditions: if no items left, no weight capacity, or no volume capacity, return 0
  if n == 0 or W_cap == 0 or V_cap == 0:
    return 0
  # If this subproblem has already been computed, return the stored result
  if memo[n][W_cap][V_cap] != -1:
    return memo[n][W_cap][V_cap]

  # Compute for the other cases
  # If the weight AND volume of the nth item can fit within the remaining capacities
  if wt[n-1] <= W_cap and vol[n-1] <= V_cap:
    # Option 1: Include the nth item
    # Add its value and recursively solve for remaining capacities and items
    include_item = val[n-1] + mem_knapSack_with_volume(wt, vol, val, W_cap - wt[n-1], V_cap - vol[n-1], n-1, memo)
    # Option 2: Exclude the nth item
    # Recursively solve for same capacities and remaining items
    exclude_item = mem_knapSack_with_volume(wt, vol, val, W_cap, V_cap, n-1, memo)
    # Store the maximum of the two options in the memoization table
    memo[n][W_cap][V_cap] = max(include_item, exclude_item)
    return memo[n][W_cap][V_cap]
  # If the nth item cannot fit (either too heavy or too bulky), it must be excluded
  else:
    # Recursively solve for same capacities and remaining items (excluding the current one)
    memo[n][W_cap][V_cap] = mem_knapSack_with_volume(wt, vol, val, W_cap, V_cap, n-1, memo)
    return memo[n][W_cap][V_cap]

# Call the function with the new criteria
result_memo_with_volume = mem_knapSack_with_volume(wt, vol, val, W_cap, V_cap, n, memo)
print(f"Max value (Memoization with max_volume): {result_memo_with_volume}")
```

Fibonacci Numbers

```python
def fibonacci_dp(n):
    # Base cases
    if n <= 1:
        return n

    # Create a DP table to store results
    dp = [0] * (n + 1)

    # Initialize base cases in the DP table
    dp[0] = 0
    dp[1] = 1

    # Fill the DP table using the recurrence relation
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

# Test the function
print("Fibonacci Series using Dynamic Programming:")
for i in range(11):
    print(f"F({i}) = {fibonacci_dp(i)}")
```

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
def fibonacci(n):
    """Find nth Fibonacci number using dynamic programming (0-indexed)."""
    if n <= 1:
        return n

    # Create DP array
    dp = [0] * (n + 1)
    dp[1] = 1

    # Fill DP array
    for i in range(2, n + 1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

# Simple test
n = 10
print(f"The {n}th Fibonacci number is: {fibonacci(n)}")
```

Problem:
You have 1GB of mobile data left. You have n apps with different data consumption rates. Choose which apps to use to maximize total value without exceeding your data limit.

```python
def choose_apps_recursive(apps, data_left, n):
    """Recursive - returns max value and which apps to use."""
    if n == 0 or data_left == 0:
        return 0, []

    data, value = apps[n-1]

    if data > data_left:
        return choose_apps_recursive(apps, data_left, n-1)

    # Use this app
    use_value, use_list = choose_apps_recursive(apps, data_left - data, n-1)
    use_value += value
    use_list = use_list + [n-1]  # Add current app index

    # Skip this app
    skip_value, skip_list = choose_apps_recursive(apps, data_left, n-1)

    if use_value > skip_value:
        return use_value, use_list
    else:
        return skip_value, skip_list

apps = [(100, 5), (300, 10), (200, 8), (50, 3), (150, 7)]
data_limit = 500
max_val, used = choose_apps_recursive(apps, data_limit, len(apps))
print(f"Recursive: Max value = {max_val}")
print(f"Used apps (indices): {sorted(used)}")
```

```python
def choose_apps_dp(apps, data_limit):
    """DP - returns max value and apps used."""
    n = len(apps)
    dp = [[0] * (data_limit + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        data, value = apps[i-1]
        for mb in range(1, data_limit + 1):
            if data > mb:
                dp[i][mb] = dp[i-1][mb]
            else:
                dp[i][mb] = max(dp[i-1][mb], value + dp[i-1][mb - data])

    # Trace back to find used apps
    used_apps = []
    mb = data_limit
    for i in range(n, 0, -1):
        if dp[i][mb] != dp[i-1][mb]:  # App was used
            used_apps.append(i-1)
            mb -= apps[i-1][0]

    total_used_data = sum(apps[i][0] for i in used_apps)

    return dp[n][data_limit], sorted(used_apps), total_used_data

# Test
apps = [(100, 5), (300, 10), (200, 8), (50, 3), (150, 7)]
data_limit = 500
max_val, used, total_data = choose_apps_dp(apps, data_limit)

print(f"\nDP Solution:")
print(f"Maximum value: {max_val}")
print(f"Used apps (indices): {used}")
print(f"Apps used:")
for i in used:
    print(f"  App {i}: {apps[i][0]}MB (value: {apps[i][1]})")
print(f"Total data used: {total_data}MB / {data_limit}MB")
```
