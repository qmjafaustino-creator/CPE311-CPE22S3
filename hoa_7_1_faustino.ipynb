## Exercise 2
1. With faang, use type conversion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker.
2. Find the seven rows with the highest value for volume.
3. Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume.


```python
import pandas as pd

fb_df = pd.read_csv('fb.csv').assign(ticker='FB')
aapl_df = pd.read_csv('aapl.csv').assign(ticker='AAPL')
amzn_df = pd.read_csv('amzn.csv').assign(ticker='AMZN')
nflx_df = pd.read_csv('nflx.csv').assign(ticker='NFLX')
goog_df = pd.read_csv('goog.csv').assign(ticker='GOOG')

faang = pd.concat([fb_df, aapl_df, amzn_df, nflx_df, goog_df], ignore_index=True)

faang['date'] = pd.to_datetime(faang['date'])
faang['volume'] = faang['volume'].astype(int)

faang = faang.sort_values(['date', 'ticker']).reset_index(drop=True)

faang
```

```python
highest_volume = faang.nlargest(7, 'volume').sort_values('volume', ascending=False)

highest_volume
```

```python
faang_long = pd.melt(
    faang,
    id_vars=['date', 'ticker'],
    value_vars=['open', 'high', 'low', 'close', 'volume'],
    var_name='variable',
    value_name='value'
)

print("Long format shape:", faang_long.shape)
print("\nFirst few rows of melted dataframe:")
faang_long.head(10)
```
