```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
print(salaries)
```

```python
mean = sum(salaries)/len(salaries)
print("mean / average is " + str(mean))
```

```python
#median
def calculate_median(salaries):
    sorted_data = sorted(salaries)
    n = len(sorted_data)
    middle = n // 2

    if n % 2 == 0:  # even number of elements
        median = (sorted_data[middle - 1] + sorted_data[middle]) / 2
    else:
        median = sorted_data[middle]

    return median


median_value = calculate_median(salaries)
print("Median is:", median_value)
```

```python
#mode
map_values = {}
for salary in salaries:
  if salary not in map_values:
    map_values[salary] = 1
    continue
  map_values[salary] += map_values[salary] + 1

print(map_values)
```

```python
#Variance
average = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1

x_xmean_squared = []
for salary in salaries:
  temp = (salary - average)**2
  x_xmean_squared.append(temp)

summation = sum(x_xmean_squared)
variance = summation/len_minus_one

print(variance)
```

```python
#standard deviation

squared_diffs = [(x - mean)**2 for x in salaries]
variance = sum(squared_diffs)/(len(salaries)-1)
std_dev = variance **0.5

print(std_dev)
```

```python
df['Classification'] = df['Diagnosis'].apply(lambda x: 'Diabetes' if x
== 1 else 'No Diabetes')
print("7. Created 'Classification' column.")
```

```python
withDiabetes = df[df['Diagnosis'] == 1].copy()
print(f"8. 'withDiabetes' dataframe created with {len(withDiabetes)}records.")
```

```python
noDiabetes = df[df['Diagnosis'] == 0].copy()
print(f"9. 'noDiabetes' dataframe created with {len(noDiabetes)}records.")
```

```python
Pedia = df[(df['Age'] >= 0) & (df['Age'] <= 19)].copy()
print(f"10. 'Pedia' (Age 0-19) dataframe created with {len(Pedia)}records.")
```
