```python
#Declaration of the "route" class which holds the attributes of the different routes
class Route:
    def __init__(self, name, profit, trips_per_day, distance, gas_used, expense):
        self.name = name
        self.profit = profit
        self.trips_per_day = trips_per_day
        self.distance = distance
        self.gas_used = gas_used
        self.expense = expense

        self.total_gas = gas_used * trips_per_day
        self.total_profit = profit * trips_per_day
        self.revenue = self.total_profit - expense
        self.value = self.revenue / self.total_gas

#Providing the data for each route
routes = [
    Route("SSS - Cubao", 3500, 8, 15, 6, 2500),
    Route("Parang - Cubao", 3800, 7, 18, 7, 3000),
    Route("Montalban - Cubao", 4800, 6, 28, 10, 4200),
    Route("Antipolo - Cubao", 4300, 8, 10, 9, 3900),
    Route("Antipolo - Angono", 5000, 5, 3, 11, 2800),
    Route("Montalban - Florida", 3000, 6, 4, 8, 2300),
    Route("Antipolo - Harvard", 4800, 6, 10.5, 12, 1850),
    Route("Antipolo - Miami", 3090, 12, 7.7, 10, 2150),
    Route("Baras - SSS", 1900, 9, 3.6, 8, 1100)
]
#constant max gas per day (constraint)
MAX_GAS = 100

#Running iteraations and the different modes
def run_iteration(routes, mode):
  #initialization and starting the different iterations from scratch
    remaining = routes.copy()
    gas_left = MAX_GAS

    chosen = []
    total_profit = 0
    total_revenue = 0
    total_gas = 0
  #While loop for program
    while gas_left > 0 and remaining:
        best = None
  #Code for optimization with constraints
        for r in remaining:
            if r.total_gas <= gas_left:
                if best is None:
                    best = r
                else:
                    if mode == "profit" and r.total_profit > best.total_profit:
                        best = r
                    elif mode == "gas" and r.total_gas > best.total_gas:
                        best = r
                    elif mode == "value" and r.value > best.value:
                        best = r

        if not best:
            break

        chosen.append(best)
        gas_left -= best.total_gas
        total_profit += best.total_profit
        total_revenue += best.revenue
        total_gas += best.total_gas
        remaining.remove(best)

    return chosen, total_profit, total_revenue, total_gas, gas_left


#Printing of outputs
all_iterations = []

for title, mode in [
    ("ITERATION 1 – MAX PROFIT", "profit"),
    ("ITERATION 2 – MAX GAS USAGE", "gas"),
    ("ITERATION 3 – MAX REVENUE PER GAS", "value")
]:
    routes_used, profit, revenue, gas_used, gas_left = run_iteration(routes, mode)
    all_iterations.append((title, routes_used, profit, revenue, gas_used, gas_left))
    print(f"\n{title}")
    print("-" * 40)

    for r in routes_used:
        print(f"• {r.name} ({r.trips_per_day} trips)")

    print("\nSummary:")
    print("Routes Used:", len(routes_used))
    print("Gas Used:", gas_used)
    print("Gas Left:", gas_left)
    print("Total Profit:", profit)
    print("Total Revenue:", revenue)

print("\nAll Routes Taken Per Iteration:")
print("-" * 50)
for title, routes_used, _, _, _, _ in all_iterations:
    print(f"\n{title}:")
    for r in routes_used:
        print(f"• {r.name} ({r.trips_per_day} trips)")
```
